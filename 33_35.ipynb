```python
import pandas as pd

df = pd.read_csv("topfit_rozdel_cviceni.csv")
df["date"] = pd.to_datetime(df["date"])
df["week"] = df["week"].astype(int)
df["year"] = df["date"].dt.year
df["year_week"] = df["year"].astype(str) + "-" + df["week"].astype(str).str.zfill(2)

print("DEBUG: ZRT rows in weeks 33–35")
display(df[(df["week"].between(33, 35)) & (df["summary_norm"] == "ZRT")])

print("DEBUG: summary_types")
summary_types = sorted(df["summary_norm"].unique())
print(summary_types)

print("DEBUG: weekly pivot")
energy_weekly = df.pivot_table(
    index="year_week",
    columns="summary_norm",
    values="energy_per_category",
    aggfunc="sum"
).reindex(sorted(df["year_week"].unique()), fill_value=0)

display(energy_weekly.loc[["2025-33", "2025-34", "2025-35"]])
```
